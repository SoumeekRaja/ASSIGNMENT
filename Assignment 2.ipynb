```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from tqdm import tqdm_notebook

from sklearn.datasets import make_blobs
```

```python
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["red","yellow","green"])
```

```python
from sklearn.datasets import make_moons, make_circles
data, labels = make_moons(n_samples=1000, random_state=0, noise=0.15)
print(data.shape, labels.shape)
plt.scatter(data[:,0], data[:,1], c=labels, cmap=my_cmap)
plt.show()
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#Single layer neural network
one_layer_model = keras.Sequential([
       layers.Dense(units=1, input_shape=[2], activation="sigmoid")
])
```

```python
one_layer_model.compile(
    loss="mse",
    optimizer="adam"
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
history_of_one_layer =  one_layer_model.fit(X_train,y_train, epochs=1000)
```

```python
import pandas as pd
```

```python
one_l_m_df = pd.DataFrame(history_of_one_layer.history)
one_l_m_df.tail()
```

```python
loss = one_l_m_df['loss']
loss = np.array(loss)
epochs_range = np.arange(1, 1001)
```

```python
plt.plot(epochs_range, loss)
plt.show()
```

```python
Y_pred_train = one_layer_model.predict(X_train)
Y_pred_binarised_train = (Y_pred_train >= 0.5).astype("int").ravel()
Y_pred_val = one_layer_model.predict(X_test)
Y_pred_binarised_val = (Y_pred_val >= 0.5).astype("int").ravel()
accuracy_train = accuracy_score(Y_pred_binarised_train, y_train)
accuracy_val = accuracy_score(Y_pred_binarised_val, y_test)
```

```python
print(accuracy_train)
print(accuracy_val)
```

```python
plt.scatter(X_train[:,0], X_train[:,1], c=Y_pred_binarised_train, cmap=my_cmap, s=15*(np.abs(Y_pred_binarised_train-y_train)+.2))
plt.show()
```

```python
#multi layer neural network
mutli_layer_model = keras.Sequential([
       layers.Dense(units=2, activation='relu', input_shape=[2]),
       layers.Dense(units=2, activation='relu'),
       layers.Dense(units=1)
])
```

```python
mutli_layer_model.compile(
    loss="mse",
    optimizer="adam"
)
```

```python
his = mutli_layer_model.fit(X_train, y_train, epochs=1000)
```

```python
mul_df = pd.DataFrame(his.history)
mul_df.tail()
```

```python
loss = mul_df['loss']
loss = np.array(loss)
epochs_range = np.arange(1, 1001)
```

```python
plt.plot(epochs_range, loss)
plt.show()
```

```python
Y_pred_train = mutli_layer_model.predict(X_train)
Y_pred_binarised_train = (Y_pred_train >= 0.5).astype("int").ravel()
Y_pred_val = mutli_layer_model.predict(X_test)
Y_pred_binarised_val = (Y_pred_val >= 0.5).astype("int").ravel()
accuracy_train = accuracy_score(Y_pred_binarised_train, y_train)
accuracy_val = accuracy_score(Y_pred_binarised_val, y_test)
```

```python
print(accuracy_train)
print(accuracy_val)
```

```python
plt.scatter(X_train[:,0], X_train[:,1], c=Y_pred_binarised_train, cmap=my_cmap, s=40*(np.abs(Y_pred_binarised_train-y_train)+.2))
plt.show()
```
